```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python

X = np.array([
    [1, 2], [2, 1], [3, 4],
    [5, 7], [3, 5], [8, 8],
    [9, 10], [8, 9], [10, 12],
    [11, 5], [12, 6], [12, 7]
])


plt.scatter(X[:, 0], X[:, 1], color='blue', s=100)
plt.title("Our Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Number of clusters
k = 3

# Initialize K-Means
kmeans = KMeans(n_clusters=k, random_state=42)

# Fit the model to our data
kmeans.fit(X)

# Get the cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Plot the points with cluster colors
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=100)

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')

plt.title("K-Means Clustering on Our Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
for i, point in enumerate(X):
    print(f"Point {point} is in Cluster {labels[i]}")
```
